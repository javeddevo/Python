```python
import pandas as pd
import sqlalchemy as sal
engine=sal.create_engine("mssql+pyodbc://LAPTOP-MRKDUIFF\SQLEXPRESS01/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER")
conn=engine.connect()
```

```python
import pandas as pd 
def extract():
    df_orders=pd.read_csv("orders.txt")
    df_returns=pd.read_csv("returns.txt")
    return df_orders,df_returns
```

```python
def transform(orders,returns):
    df=pd.merge(orders,returns,how="inner",left_on="order_id",right_on="order_id")
    return df

    
```

```python
#load
def load(df):
    df.to_sql("orders_final",con=conn,index=False,if_exists="replace")
    conn.commit()
    
```

```python
#extract
df_orders,df_returns=extract()
#transform
df=transform(df_orders,df_returns)
#load
load(df)
```
